```python
import os
import torch
import torchvision.utils as vutils

# Path to your trained Generator checkpoint (epoch 15)
ckpt_path = r".\out_celeba_run2\netG_epoch_15.pth"

assert os.path.exists(ckpt_path), f"Checkpoint not found: {ckpt_path}"

# Reload netG weights (keeps same architecture as defined above)
netG.load_state_dict(torch.load(ckpt_path, map_location=device))
netG.eval()

# Generate a batch of fake images
n_images = 64
noise = torch.randn(n_images, 100, 1, 1, device=device)
with torch.no_grad():
    fake = netG(noise).detach().cpu()

# Save a grid image (normalized to [0,1] for viewing)
out_path = "generated_epoch15.png"
vutils.save_image(fake, out_path, normalize=True, nrow=8)
out_path
```
